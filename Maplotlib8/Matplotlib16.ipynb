```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

```python
data = {'Item A':109438,
       'Item B':103569,
       'Item C':112214,
       'Item D':112591,
       'Item E':100934,
       'Item F':103660,
       'Item G':137351,
       'Item H':123381,
       'Item I':135841,
       'Item J':104437}
data
```

```python
items = tuple(data.keys())
items
```

```python
count = tuple(data.values())
count
```

```python
flg,ax = plt.subplots()

ax.barh(items, count)

plt.show()
```

```python
flg,ax = plt.subplots()
ax.barh(items, count)

labels=ax.get_xticklabels()

plt.setp(labels,
       rotation=45,
       horizontalalignment='right')

plt.show()
```
